```python
import matplotlib.pyplot as plt

from config import load_config
from rsna_dataloader import RSNAPneumoniaDataModule

CONFIG_PATH = "config.yml"
NUM_IMAGES = 8
```

```python
configs = load_config(CONFIG_PATH)
rsna = RSNAPneumoniaDataModule(configs, batch_size=NUM_IMAGES)
test_data_batch = rsna.test_dataloader()
test_data = next(iter(test_data_batch))
```

```python
fig, ax = plt.subplots(2, NUM_IMAGES//2, figsize=(NUM_IMAGES*2, 8))
idx = 0
for r in range(2):
    for c in range(NUM_IMAGES//2):
        img = test_data["image"][idx].squeeze().permute(1,2,0).int()
        ax[r][c].set_title("Label: " + str(test_data["target"][idx].item()))
        ax[r][c].imshow(img, cmap="gray")
        ax[r][c].axis("off")
        idx+=1
```
